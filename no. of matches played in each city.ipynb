```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
matches=pd.read_csv("Ipl Data/matches.csv")
```

```python
plt.figure(figsize = (30, 20))
sns.countplot(x='city',data=matches)
plt.title('No of matches played in each season',fontsize=25)
plt.xticks(rotation = 50, ha = 'right', fontsize = 18)
plt.yticks(fontsize=15)
plt.xlabel('Season',fontsize=22)
plt.ylabel('No. of matches',fontsize=22)
plt.show()
```
